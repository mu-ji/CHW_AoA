```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file_name = '0000_72_0.5_90_90_1.npy'
antenna_switch_data = np.load('{}'.format(data_file_name))

fig, ax = plt.subplots(figsize=(10, 6))
antenna_switch_data_mean = np.mean(antenna_switch_data, axis=2)
plt.violinplot(antenna_switch_data_mean, showmeans=True)
plt.xlabel('Antenna switch id')
plt.ylabel('Phase difference')
plt.show()
```

```python
data_file_name = '0005_72_0.5_90_90_1.npy'
antenna_switch_data = np.load('{}'.format(data_file_name))

def release_jump(diff):
    for i in range(len(diff)):
        if abs(diff[i]) > 201:
            if diff[i] < -201:
                diff[i] = diff[i] + 402
            if diff[i] > 201:
                diff[i] = diff[i] - 402
    return diff

phase_diff_data = np.zeros((100,15,8))
for i in range(100):
    phase_diff = antenna_switch_data[i,1:,:] - antenna_switch_data[i,0,:]
    for j in range(15):
        phase_diff[j] = release_jump(phase_diff[j])
    
    phase_diff_data[i] = phase_diff

fig, ax = plt.subplots(figsize=(10, 6))
antenna_switch_data_mean = np.mean(phase_diff_data, axis=2)
print(antenna_switch_data_mean[20,:])
plt.violinplot(antenna_switch_data_mean, showmeans=True)
plt.xlabel('Switch times for same antenna')
plt.ylabel('Phase difference')
plt.show()
```
